```python
import pandas as pd

tabela = pd.read_csv("PDA_Dados_Cursos_Graduacao_Brasil.csv")
display(tabela)
```

Análise da Distribuição das Instituições de Ensino por Região

```python
tabela = tabela[tabela['REGIAO'] != 'IGNORADO/EXTERIOR']
contagem_regioes = tabela['REGIAO'].value_counts()
display(contagem_regioes)
```

Análise da Distribuição das Instituições de Ensino por Região - Gráfico

```python
import matplotlib.pyplot as plt

tabela = tabela[tabela['REGIAO'] != 'IGNORADO/EXTERIOR']

contagem_regioes = tabela['REGIAO'].value_counts()

plt.figure(figsize=(10, 6))
contagem_regioes.plot(kind='bar', color='skyblue')
plt.title('Distribuição das Instituições de Ensino por Região')
plt.xlabel('Região')
plt.ylabel('Número de Instituições')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Distribuição de Cursos por Modalidade

```python
tabela = tabela.dropna(subset=['MODALIDADE'])
contagem_modalidade = tabela['MODALIDADE'].value_counts()
display(contagem_modalidade)
```

Distribuição de Cursos por Modalidade - Gráfico

```python
import matplotlib.pyplot as plt

tabela = tabela.dropna(subset=['MODALIDADE'])

contagem_modalidade = tabela['MODALIDADE'].value_counts()

plt.figure(figsize=(8, 8))
contagem_modalidade.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Distribuição de Cursos por Modalidade')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

Distribuição de Vagas Autorizadas por Curso (Top 10)

```python
vagas_por_curso = tabela.groupby('NOME_CURSO')['QT_VAGAS_AUTORIZADAS'].sum()
vagas_por_curso = vagas_por_curso.sort_values(ascending=False)
display(vagas_por_curso)
```

Distribuição de Vagas Autorizadas por Curso (Top 10) - Gráfico

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
vagas_por_curso.head(10).plot(kind='bar', color='skyblue')
plt.title('Distribuição de Vagas Autorizadas por Curso (Top 10)')
plt.xlabel('Curso')
plt.ylabel('Quantidade de Vagas Autorizadas')

plt.tight_layout()
plt.show()
```
